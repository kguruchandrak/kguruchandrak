```python
# Program to check if the given number is a Disarium Number?

num = int(input("Enter number : "))
rem = s = 0;    
l = len(str(num))

n = num;    
     
while(num > 0):    
    rem = num%10;    
    s += int(rem**l);    
    num = num//10;    
    l -= 1;    
     
if(s == n):    
    print("{} is a Disarium number".format(n));    
else:    
    print("{} is not a disarium number".format(n)); 
```

```python
# Program to print all disarium numbers between 1 to 100?

for num in range(1,101):
    rem = s = 0;    
    l = len(str(num))
    n = num;    
     
    while(num > 0):    
        rem = num%10;    
        s += int(rem**l);    
        num = num//10;    
        l -= 1;    
     
    if(s == n):    
        print("{} is a Disarium number".format(n));    
    else:    
        print("{} is not a disarium number".format(n)); 
```

```python
# Program to check if the given number is Happy Number?

def isHappy(n):    
    r = s = 0;    
    while(n > 0):    
        r = n%10    
        s += r**2  
        n //= 10    
    return s  
        
num = int(input("Enter num : "))    
res = num
while(res != 1 and res != 4):    
    res = isHappy(res)    
     
if(res == 1):    
    print("{} is a Happy number".format(num));    
else:    
    print("{} is not a Happy number".format(num)); 
```

```python
# Program to print all happy numbers between 1 and 100?

def isHappy(n):    
    r = s = 0;    
    while(n > 0):    
        r = n%10    
        s += r**2  
        n //= 10    
    return s  

for num in range(1,101):       
    res = num
    while(res != 1 and res != 4):    
        res = isHappy(res)    
     
    if(res == 1):    
        print("{} is a Happy number".format(num));    
    else:    
        print("{} is not a Happy number".format(num)); 
```

```python
# Program to determine whether the given number is a Harshad Number?

num = int(input("Enter num : "))
copy = num
digit_sum = 0

while num:
    digit_sum += num%10
    num //= 10
    
if copy%digit_sum == 0:
    print("{} is Harshad Number".format(copy))
else:
    print("{} is not a Harshad Number".format(copy))
```

```python
# Program to print all pronic numbers between 1 and 100?

def PronicNum(num):    
    flag = False;    
        
    for j in range(1, num+1):      
        if((j*(j+1)) == num):    
            flag = True;    
            break;    
    return flag;    

print("Pronic numbers between 1 and 100 : ");    
for i in range(1, 101):    
    if(PronicNum(i)):    
        print(i,end = " ")    
        
```
